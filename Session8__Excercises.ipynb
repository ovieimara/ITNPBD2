## 8.4 Matplotlib

In an imaginary scenario, a parent has approached you and asked you to visualise their child's growth, from 1 to 10 years old. Their heights each year were: 76cm, 87cm, 95cm, 102cm, 109cm, 115cm, 122cm, 128cm, 133cm, 138cm.

Plot this as a line graph, ensuring that the X and Y axis are labeled appropriately, a title is included, and with X starting at 1.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(np.arange(1, 11), [76, 87, 95, 102, 109, 115, 122, 128, 133, 138], 'r-')
plt.xticks(range(0, 11))
plt.xlabel('Years')
plt.ylabel('Height (cm)')
plt.title('Child Growth')
```

Add another child to your graph that had the heights 70cm, 84cm, 92cm, 100cm, 107cm, 115cm, 124cm, 134cm, 143cm, 147cm. Ensure that you include a legend to distinguish between the children.

```python
plt.plot(np.arange(1, 11), [76, 87, 95, 102, 109, 115, 122, 128, 133, 138], 'r-')
plt.plot(np.arange(1, 11), [ 70, 84, 92, 100, 107, 115, 124, 134, 143, 147])
plt.legend(['Child 1', 'Child 2'])
plt.xticks(range(0, 11))
plt.xlabel('Years')
plt.ylabel('Height (cm)')
plt.title('Child Growth')
plt.show()
```

Change the font and fontsize of the axis labels, and make the lines green and black.

```python
plt.plot(np.arange(1, 11), [76, 87, 95, 102, 109, 115, 122, 128, 133, 138], 'g-')
plt.plot(np.arange(1, 11), [ 70, 84, 92, 100, 107, 115, 124, 134, 143, 147], 'black')
plt.legend(['Child 1', 'Child 2'])
plt.xticks(range(0, 11))
plt.xlabel('Years', fontdict={'family': 'fantasy', 'size' : 12})
plt.ylabel('Height (cm)', fontdict={'family': 'fantasy', 'size' : 12})
plt.title('Child Growth', fontdict={'family': 'fantasy', 'size' : 12})
plt.show()
```
